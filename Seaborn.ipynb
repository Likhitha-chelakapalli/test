# Lineplot
A line plot is a type of data visualization used to display trends over a continuous variable, typically time or some ordered numeric variable

### Synatx
import seaborn as sns

import matplotlib.pyplot as plt

**sns.lineplot(dta=None, x=None, y=None, hue=None, style=None, markers=False, dashes=True)**

plt.show()

| Parameter | Description                                                     |
| --------- | --------------------------------------------------------------- |
| `data`    | Dataset (e.g., DataFrame or dictionary)                         |
| `x`       | Variable for x-axis (usually time or index)                     |
| `y`       | Variable for y-axis (value to be plotted)                       |
| `hue`     | Grouping variable that will produce lines with different colors |
| `style`   | Grouping variable that will produce different line styles       |
| `markers` | Add markers at data points (True/False or list of styles)       |
| `dashes`  | Line dashes (True/False or list of styles)                      |


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Dataset with categories
data = pd.DataFrame({
    'day': [1, 2, 3, 1, 2, 3],
    'temperature': [30, 32, 31, 28, 29, 27],
    'city': ['A', 'A', 'A', 'B', 'B', 'B']
})

# Plot with hue
sns.lineplot(data=data, x='day', y='temperature')
plt.title("Temperature Trend in Two Cities")
plt.show()
```

```python
# Dataset with categories
data = pd.DataFrame({
    'day': [1, 2, 3, 1, 2, 3],
    'temperature': [30, 32, 31, 28, 29, 27],
    'city': ['A', 'A', 'A', 'B', 'B', 'B']
})

# Plot with hue
sns.lineplot(data=data, x='day', y='temperature', hue='city', style='city',markers=True)
plt.title("Temperature Trend in Two Cities")
plt.show()
```

##  What is relplot()?
relplot() is a wrapper for scatterplot() and lineplot().

It is used when you want to create multiple subplots using facets (e.g., using col, row, or hue to divide the dataset).

The actual kind of plot drawn is controlled by the kind parameter: either 'scatter' or 'line'.
 

| Parameter    | Description                                                   |
| ------------ | ------------------------------------------------------------- |
| `data`       | Dataset (e.g., Pandas DataFrame)                              |
| `x`, `y`     | Variables to plot on the X and Y axes                         |
| `kind`       | `'scatter'` or `'line'`                                       |
| `hue`        | Color grouping                                                |
| `style`      | Line/dot style grouping                                       |
| `size`       | Marker size (for scatter plots)                               |
| `col`, `row` | Create subplots by columns/rows based on categorical variable |
| `height`     | Height (in inches) of each subplot                            |
| `aspect`     | Aspect ratio (width/height) of each facet                     |


```python
tips = sns.load_dataset("tips")
sns.relplot(data=tips, x="total_bill", y="tip",kind='scatter')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = sns.load_dataset("tips")
sns.relplot(
    data=df,
    x="size", y="tip",
    kind="line",
    hue="sex",          # Color by sex
    col="smoker",      # Facet by time (Lunch/Dinner) Create the different plots based on the unique values in the 'time' column value
    style="time",     # Different line styles
    markers=True,
)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = sns.load_dataset("tips")
sns.relplot(
    data=df,
    x="size", y="tip",
    kind="line",
    hue="sex",          # Color by sex
    col="smoker",      # Facet by time (Lunch/Dinner) Create the different plots based on the unique values in the 'time' column value
    style="time",     # Different line styles
    markers=True,
    errorbar=None # used to remove shaded area
)
plt.show()
```

# **Seaborn.displot()**

1. seaborn.displot() is a figure-level function in the Seaborn library used for visualizing the distribution of a dataset.
2. It can create histograms, kernel density plots (KDEs), and empirical cumulative distribution functions (ECDFs).
3. This function is flexible and combines the features of several axis-level functions, such as sns.histplot(), sns.kdeplot(), and sns.ecdfplot().
4. **seaborn.displot(
    data=None, x=None, y=None, hue=None, row=None, col=None,
    weights=None, kind='hist', rug=False, rug_kws=None, log_scale=None,
    height=5, aspect=1, facet_kws=None)**


| Parameter           | Purpose                                | Example                       |
| ------------------- | -------------------------------------- | ----------------------------- |
| `x`, `y`            | Variables to plot                      | `x="body_mass_g"`             |
| `data`              | Source data                            | `data=df`                     |
| `kind`              | Type of plot (`hist`, `kde`, `ecdf`)   | `kind="kde"`                  |
| `hue`, `row`, `col` | Grouping variable                      | `hue="species"`               |
| `weights`           | Weight data points                     | `weights="weights"`           |
| `rug`               | Add rug lines                          | `rug=True`                    |
| `log_scale`         | Use log scale                          | `log_scale=True`              |
| `height`, `aspect`  | Control plot size                      | `height=4, aspect=1.2`        |
| `facet_kws`         | Grid layout tweaks                     | `facet_kws={"sharey": False}` |
| `**kwargs`          | Style/histogram options (e.g., `bins`) | `bins=20, color='blue'`       |


**1.kind:
Type: {"hist", "kde", "ecdf"}
Purpose: Specifies type of distribution plot.**

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("penguins")
sns.displot(data=df, x="flipper_length_mm",kind='hist') #default is hist
df
```

# **seaborn.kdeplot()**

1. seaborn.kdeplot() is an axis-level function used to plot Kernel Density Estimation (KDE), which is a smoothed version of a histogram used to estimate the probability density function (PDF) of a continuous variable.
2. KDE is useful when you want to see the distribution shape without bins.

**seaborn.kdeplot(
    data=None, x=None, y=None, hue=None, weights=None, common_norm=True,
    common_grid=False, cumulative=False, bw_method=None, bw_adjust=1,
    fill=False, vertical=False, clip=None, cut=3, gridsize=200, 
    kernel='gau', bw=None, log_scale=False, color=None, palette=None,
    warn_singular=True, ax=None)**

| Parameter         | Functionality           | Example                             |
| ----------------- | ----------------------- | ----------------------------------- |
| `x`, `y`          | Data to plot            | `x="flipper_length_mm"`             |
| `hue`             | Color by category       | `hue="species"`                     |
| `fill`            | Fill under the curve    | `fill=True`                         |
| `bw_adjust`       | Adjust smoothness       | `bw_adjust=0.5`                     |
| `cumulative`      | Show CDF instead of PDF | `cumulative=True`                   |
| `clip`            | Trim curve to range     | `clip=(170, 230)`                   |
| `cut`             | Limit beyond data       | `cut=0`                             |
| `gridsize`        | Resolution of curve     | `gridsize=300`                      |
| `log_scale`       | Use log scale           | `log_scale=True`                    |
| `color`/`palette` | Set colors              | `color='blue'` / `palette='pastel'` |


**1.common_norm: Normalize across all subsets (True) or within each (False).Use: Set to False when comparing category-wise true densities.**

```python
sns.kdeplot(data=df, x="flipper_length_mm", common_norm=True)
```

**cumulative:
Plot cumulative distribution function (CDF).
Use: To see probability accumulation.**

```python
sns.kdeplot(data=df, x="flipper_length_mm", hue="species", cumulative=True)
```

**clip:
Set bounds beyond which KDE is not plotted.**

```python
sns.kdeplot(data=df, x="flipper_length_mm", clip=(170, 230))
```

** 13. kernel:
Choose the type of kernel:
'gau' (default) for Gaussian,
'cos', 'biw', 'epa', etc.
 Rarely changed. Keep 'gau' unless experimenting.**

```python
sns.kdeplot(data=df, x="flipper_length_mm", color='pink')
```

```python
sns.kdeplot(data=df, x="flipper_length_mm",hue='species', warn_singular=True)
```

# **regplot()**

regplot() creates a scatter plot with a regression line.

It combines raw data points and a fitted linear model.

It's best for small to medium-sized datasets when you want to visualize trends, correlation, or linear relationships.

seaborn.regplot(data=None, *, x=None, y=None, x_estimator=None, x_bins=None, x_ci='ci', scatter=True, fit_reg=True, ci=95, n_boot=1000, units=None, seed=None, order=1, logistic=False, lowess=False, robust=False, logx=False, x_partial=None, y_partial=None, truncate=True, dropna=True, x_jitter=None, y_jitter=None, label=None, color=None, marker='o', scatter_kws=None, line_kws=None, ax=None)

| Parameter        | Description               | Example                     |
| ---------------- | ------------------------- | --------------------------- |
| `x`, `y`, `data` | Input data                | `x="total_bill", y="tip"`   |
| `scatter`        | Show/hide dots            | `scatter=False`             |
| `fit_reg`        | Show/hide regression line | `fit_reg=False`             |
| `ci`             | Confidence interval       | `ci=95`                     |
| `order`          | Polynomial order          | `order=2`                   |
| `lowess`         | Non-linear smooth curve   | `lowess=True`               |
| `robust`         | Robust linear fit         | `robust=True`               |
| `x_bins`         | Discretize x values       | `x_bins=4`                  |
| `x_estimator`    | Aggregate repeated x's    | `x_estimator=np.mean`       |
| `line_kws`       | Style regression line     | `line_kws={'color':'r'}`    |
| `scatter_kws`    | Style data points         | `scatter_kws={'alpha':0.5}` |


```python

# Load data
tips = sns.load_dataset("tips")
```

```python
sns.regplot(data=tips, x="total_bill", y="tip")
```

**scatter, fit_reg**

scatter: Show/hide scatter points.

# **pairplot**

pairplot() creates a grid of scatter plots for all pairwise combinations of numerical features in a DataFrame.

Also shows histograms (or KDE plots) on the diagonal.

Helps visualize correlations, distributions, and clusters in multi-dimensional data.



| Parameter   | Description                | Example                |
| ----------- | -------------------------- | ---------------------- |
| `hue`       | Color by category          | `hue="species"`        |
| `vars`      | Only specific columns      | `vars=["x1", "x2"]`    |
| `kind`      | Type of plot in grid       | `"scatter"` or `"kde"` |
| `diag_kind` | Type of diagonal plot      | `"hist"` or `"kde"`    |
| `corner`    | Show only lower triangle   | `corner=True`          |
| `plot_kws`  | Control scatter plot style | `{"alpha": 0.6}`       |


```python
penguins = sns.load_dataset("penguins")
sns.pairplot(penguins)
```

```python
# its possible to force marginal histograms:

sns.pairplot(penguins, hue="species", diag_kind="hist")
```

```python
iris = sns.load_dataset("iris")
```

```python
# vars
# Subset of numeric variables to plot.
sns.pairplot(data=iris, vars=["sepal_length", "petal_length"])
```

```python
# 4. x_vars, y_vars
# Define specific X and Y variable sets (for asymmetric grids).
sns.pairplot(data=iris, x_vars=["sepal_length"], y_vars=["petal_length", "sepal_width"])
```

```python
# 5. kind
# Type of plot in off-diagonal: 'scatter' or 'kde'.
sns.pairplot(data=iris, hue="species", kind="kde") # default scatter
```

```python
#  6. diag_kind
# Plot on the diagonal: 'hist', 'kde', or 'auto'.


sns.pairplot(data=iris, diag_kind="kde")
```

## seaborn.heatmap()

heatmap() plots a 2D matrix using colors to represent values.

Useful for:

Correlation matrices 

Confusion matrices 

Feature similarity heatmaps 

Pivot tables 

| Parameter    | Purpose                       | Example                  |
| ------------ | ----------------------------- | ------------------------ |
| `annot`      | Show values in cells          | `annot=True`             |
| `fmt`        | Format string for annotations | `fmt=".2f"`              |
| `cmap`       | Color scheme                  | `"coolwarm"`, `"YlGnBu"` |
| `center`     | Central value for colormap    | `center=0`               |
| `linewidths` | Space between cells           | `linewidths=0.5`         |
| `mask`       | Hide part of matrix           | `np.triu(...)`           |
| `square`     | Equal cell height and width   | `square=True`            |


```python
# Pass a DataFrame to plot with indices as row/column labels:

glue = sns.load_dataset("glue").pivot(index="Model", columns="Task", values="Score")
sns.heatmap(glue)
```

```python
# Use annot to represent the cell values with text:

sns.heatmap(glue, annot=True)
```

```python
sns.heatmap(data=iris.select_dtypes(include='number'),annot=True)
```

```python
# control the annotations with a formatting string:

sns.heatmap(glue, annot=True, fmt=".1f")
```

```python
# Add lines between cells:

sns.heatmap(glue, annot=True, linewidth=1.5)
```
